### Matplot library tutorial


```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("a scatter graph eg")
# to save the graph pic use plt.savefig('test.png') 
```

```python
y=x*x
print(x)
print(y)
```

```python
plt.plot(x,y,'c.--')
```

```python
# subplots
plt.subplot(2,2,1)
plt.plot(x,y,'ro')
plt.subplot(2,2,2)
plt.plot(x,y,'m*')
plt.subplot(2,2,3)
plt.plot(x,y,'b.')
plt.subplot(2,2,4)
plt.plot(x,y,'g^')
```

```python
# equation graphs 
x = np.arange(1,11)
y = (3*x)+5
plt.plot(x,y,'k.--')
```

```python
# trigo functions graphs
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y,'r-')
plt.title("sine wave graph")
```

```python
# bar graph 
x = np.arange(0,10,2)
y = np.arange(11,20,2)
x2 = x*2
y2 = y*3
plt.bar(x,y, color = 'y')
plt.show()
```

```python
#histograms 
a = np.random.randint(101, size = 20)
plt.hist(a, bins = 20)
plt.title('histogram')
```

```python
# piecharts
labels = "mango", "chocolate", "butterscotch", "vanila"
sizes = [50, 20, 33, 14]
explode = (0.07, 0, 0, 0)
colors = ['blue', 'green', 'yellow', 'red']
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = True)
plt.axis('equal')
plt.show()
```
