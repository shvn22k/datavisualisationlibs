```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.dtypes
```

```python
corr_mat = df.corr(numeric_only=True)
corr_mat
```

```python
sns.heatmap(corr_mat, annot=True)
```

## Bivariate Analysis - jointplot

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='hex') # color = any matplotlib color - optional
```

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='kde', color='green') #kernel density estimation 
```

```python
sns.jointplot(x='tip',y='total_bill', data=df, kind='reg', color = 'red') 
#gives a best fit regression line and a probability density function line too
```

## Multivariate Analysis - pairplot and distplot (for one by one feature kde)

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex', palette = 'magma')
```

```python
sns.pairplot(df, hue = 'smoker', palette = 'mako')
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'], kde = False, bins = 10)
```
